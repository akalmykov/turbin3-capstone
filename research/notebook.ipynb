```python
from solana.rpc.api import Client
import matplotlib.pyplot as plt
from tqdm import tqdm
import json
import time

client = Client("https://api.mainnet-beta.solana.com")

current_slot = json.loads(client.get_slot().to_json())["result"]
slots = list(range(current_slot, current_slot - 1000, -1))

timestamps = []
for slot in tqdm(slots):
    resp = client.get_block_time(slot)
    ts = json.loads(resp.to_json()).get("result")
    if ts:
        timestamps.append(ts)
    time.sleep(1)
    if slot % 100 == 0:
        with open("slot_data.json", "w") as f:
            json.dump(
                {
                    "current_slot": current_slot,
                    "slots_processed": slot,
                    "timestamps": timestamps,
                },
                f,
            )
```

```python
import numpy as np

# Read timestamps from file
with open("slot_data.json", "r") as f:
    data = json.load(f)
    timestamps = data["timestamps"]
print("timestamps length: ", len(timestamps))

consecutive_timestamps = list(zip(timestamps[:-1], timestamps[1:]))
print("consecutive_timestamps: ", consecutive_timestamps)
intervals = [t1 - t2 for t1, t2 in consecutive_timestamps]

print("intervals: ", intervals)
print("intervals length: ", len(intervals))
print("mean interval: ", sum(intervals) / len(intervals))
print("median interval: ", sorted(intervals)[len(intervals) // 2])
print("min interval: ", min(intervals))
print("max interval: ", max(intervals))
print("std interval: ", np.std(intervals))
```
